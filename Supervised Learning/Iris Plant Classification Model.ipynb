This Jupyter Notebook showcases a classification model designed to predict the species of the Iris plant using a well-known dataset from the field of machine learning. The dataset consists of 50 samples from each of the three species of Iris (Iris setosa, Iris versicolor, and Iris virginica), with measurements of the petal and sepal lengths and widths.

In this project, I utilize the following Python libraries:

**Any of the lines would serve for our model since they separate the data (remember that in most cases we won't be talking about just 2 dimensions)**
<br><br>
**We can chose the <font color="red"> red </font> because it is neither too close to 0 nor too close to 1.**

**This line is given by the equation below.:**<br>
**<font color="red"> y = -x + 3** </font>

**To determine in which region any point will be, we just need to substitute the x-coordinate of the point into the equation above and check.:**<br><br>
- If the y-coordinate of the point is greater than the y-coordinate of the equation, this point will be classified as 1.
- If the y-coordinate of the point is less than the y-coordinate of the equation, this point will be classified as 0.

**Ex: (3.5,0.7)**
- y from the equation = -x + 3 = -3.5 + 3 = -0.5
- point y = 0.7
<br><br>
- In this case: **`y of the point > y of the equation`** -> **Target = 1**


**To visualize, we can plot this point on the graph above.**

```python
# Drawing the line again and adding the point above
fig, ax = plt.subplots()

# Here we are making a scatter plot
x = iris["petal length (cm)"]
y = iris["petal width (cm)"]
ax.scatter(x,y,c=iris.Target,cmap='PiYG')

# Here we are making a line graph
x_straight3 = np.arange(1,6)
y_straight3 = (-1)*x_straight3+3
ax.plot(x_straight3,y_straight3,c='r')

# Making the scatterplot only from this point
ax.scatter(2,+0.6,c='r')

# Making the scatterplot only from this point
ax.scatter(2,+1,c='b')

plt.show()
```

```python
def classify_model(x,y):
     y_equation = -x + 3
     y_point = y
    
     if y_point > y_equation:
         return 1
     elif y_point < y_equation:
         return 0
     else:
         return -1
```

```python
# (3.5,0.7)
classify_model(3.5,0.7)
```

```python
# (2,0.6)
classify_model(2,0.6)
```

# Creating a Machine Learning model to classify my points

### Let's use Perceptron to find the best straight line that separates these points
- https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Perceptron.html#sklearn.linear_model.Perceptron

```python
display(Image(filename='perceptron.png'))
```

```python
# Importing the perceptron
from sklearn.linear_model import Perceptron
clf = Perceptron(tol=1e-3, random_state=0)
```

```python
# Using for our data
iris.head(2)
```

```python
# Selecting the X and the y

X = iris[['petal length (cm)','petal width (cm)']]
y = iris.Target
```

```python
# Fitting the model
clf.fit(X, y)
```

```python
# w1 e w2
clf.coef_
```

```python
# w0
clf.intercept_
```

w0 + w1.x + w2.y = 0

w2.y = - w0 - w1.x

y = (- clf.intercept_ - clf.coef_[0][0].x)/clf.coef_[0][1]

```python
clf.coef_[0][1]
```

```python
# Plotting both the perceptron and the line we drew on the graph
```

```python
# Plotting the line again and adding the point above
fig, ax = plt.subplots()

# Here we are making a scatter plot
x = iris["petal length (cm)"]
y = iris["petal width (cm)"]
ax.scatter(x,y,c=iris.Target,cmap='PiYG')

# Here we are making a line graph
x_reta3 = np.arange(1,6)
y_reta3 = (-1)*x_reta3+3
ax.plot(x_reta3,y_reta3,c='r')

# Here we are creating a line chart.
x_perc = np.arange(1,6)
y_perc = (-clf.intercept_-clf.coef_[0][0]*x_perc)/clf.coef_[0][1]
ax.plot(x_perc,y_perc,c='#FF5733')

plt.show()
```

**Adding a new dataset with the data shown below**
- These data were created by us and serve only to show what would happen in case of adding new data to the model

```python
# Adding a new dataset with the data shown below
base2 = pd.DataFrame([[0,0,3,0.75,0], [0,0,2.5,1,0]])
base2.columns = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','Target']
```

```python
base2.head()
```

```python
# The given data has been added to the iris dataset.
iris = iris.append(base2)
```

```python
iris.tail()
```

```python
# Running the Perceptron for this new dataset.
X = iris[["petal length (cm)","petal width (cm)"]]
y = iris.Target

clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(X, y)
```

**Plotting the graph again with the perceptron and our line for the new data**

```python
# Plotting the line again and adding the point above
fig, ax = plt.subplots()

# Here we are creating a scatter plot
x = iris["petal length (cm)"]
y = iris["petal width (cm)"]
ax.scatter(x,y,c=iris.Target,cmap='PiYG')

# Here we are making a line graph
x_reta3 = np.arange(1,6)
y_reta3 = (-1)*x_reta3+3
ax.plot(x_reta3,y_reta3,c='r')

# Here we are creating a line graph
x_perc = np.arange(1,6)
y_perc = (-clf.intercept_-clf.coef_[0][0]*x_perc)/clf.coef_[0][1]
ax.plot(x_perc,y_perc,c='#FF5733')

plt.show()
```

**Note that now, for the new data, the perceptron can adjust itself and draw a new line, continuing to classify all points correctly. However, in our initially drawn line, these two new points would be classified incorrectly.**
